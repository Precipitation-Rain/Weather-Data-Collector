# How to access data from API using RestApi

```python
import requests    # to make HTTP requests
import pandas as pd  # to convert JSON to table and save as CSV

# ──────────────────────────────────────────
# ENDPOINT — Main URL
# this is where CoinGecko keeps coins market data
# ──────────────────────────────────────────
url = "https://api.coingecko.com/api/v3/coins/markets"

# ──────────────────────────────────────────
# QUERY PARAMETERS — Special Instructions
# telling the server exactly what we want
# instead of putting them in URL we pass as dictionary
# requests library automatically adds ? and & for us
# ──────────────────────────────────────────
params = {
    "vs_currency": "usd",          # give prices in USD
    "order": "market_cap_desc",    # sort by market cap (biggest first)
    "per_page": 100,               # give 100 coins
    "page": 1,                     # first page
    "sparkline": False             # no sparkline data (keep it simple)
}

# ──────────────────────────────────────────
# REQUEST — One direct call, no session needed
# no homepage visit, no cookies, no cloudscraper
# CoinGecko is open — one shot directly
# params are automatically added to URL by requests
# ──────────────────────────────────────────
response = requests.get(url, params=params)
print(f"Status code: {response.status_code}")
# ✅ Should be 200

# ──────────────────────────────────────────
# SEE FULL URL — with params attached
# so you can see what requests built automatically
# ──────────────────────────────────────────
print(f"Full URL called: {response.url}")

# ──────────────────────────────────────────
# PARSE JSON → DataFrame
# response.json() converts raw response to Python list
# each item in list = one coin = one row
# ──────────────────────────────────────────
data = response.json()
print(f"Type of data: {type(data)}")        # should be list
print(f"Total coins received: {len(data)}") # should be 100
print(f"First coin: {data[0]}")             # see raw JSON of first coin

# ──────────────────────────────────────────
# CONVERT TO DATAFRAME → CSV
# ──────────────────────────────────────────
df = pd.DataFrame(data)
print(df.head())                      # preview first 5 rows
print(df.columns.tolist())            # see all column names

df.to_csv("crypto_market_data.csv", index=False)
print("✅ Saved to crypto_market_data.csv")
```

# How to access Data from all pages 

```python
import time
import pandas as pd
import requests
```

```python
page = 1
all_data = []
url = "https://api.coingecko.com/api/v3/coins/markets"
```

```python
while True:

    params = {
    "vs_currency": "usd",          # give prices in USD
    "order": "market_cap_desc",    # sort by market cap (biggest first)
    "per_page": 250,               # give 100 coins
    "page": page,                     # first page
    "sparkline": False             # no sparkline data (keep it simple)
    }

    response = requests.get(url , params=params)
    print("Response status" ,response.status_code)
    response_code = response.status_code

    if response_code == 429:
        print("We are hitting too fast")
        time.sleep(30)
        continue
    
    if response_code != 200:
        print("Errorr Error Errror")
        break

    data = response.json()
    print(f" Page{page} -> Data Received {len(data)}" )

    if len(data) == 0:
        print("Data is finished")
        break

    all_data.extend(data)
    page = page + 1
    time.sleep(2)
```

```python
temp = all_data
```

```python
len(temp)
```

```python
df = pd.DataFrame(temp)
```

```python
df.to_csv("Cryptocurrency Data.csv")
```

# Open-meteo Weather Data

## Satara City Daily Data

```python
import requests
import pandas as pd

url = "https://archive-api.open-meteo.com/v1/archive"

params = {
    "latitude":  17.68589,
    "longitude": 73.99333,
    "start_date": "2025-01-01",
    "end_date": "2025-12-31",
    "daily": [
        "weather_code",
        "temperature_2m_max",
        "temperature_2m_min",
        "apparent_temperature_max",
        "apparent_temperature_min",
        "sunrise",
        "sunset",
        "daylight_duration",
        "sunshine_duration",
        "precipitation_sum",
        "rain_sum",
        "snowfall_sum",
        "precipitation_hours",
        "wind_speed_10m_max",
        "wind_gusts_10m_max",
        "wind_direction_10m_dominant",
        "shortwave_radiation_sum",
        "et0_fao_evapotranspiration"
    ],
    "timezone": "Asia/Kolkata"
}

response = requests.get(url, params=params)
print(f"Status: {response.status_code}")

if response.status_code != 200:
    print("Error")
    print(response.text)

data = response.json()

daily_data = data["daily"]

df = pd.DataFrame(daily_data)
# print(f"Total records: {len(df)}")
# print(f"Total columns: {len(df.columns)}")
# print(df.head())

df.to_csv("Satara_Weather_Data.csv", index=False)
print("Satara_Weather_Data.csv")
```
